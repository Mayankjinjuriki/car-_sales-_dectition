```python
# Imputing the null values
for i in df.columns:
    if df[i].dtype == 'object':
        df.fillna(df[i].mode(), inplace = True)
    else:
        df.fillna(df[i].median(), inplace = True)
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
def outlier_detection_treatment(col):
    q1= df[col].quantile(0.25)
    q3= df[col].quantile(0.75)
    iqr= q3-q1
    lf= q1- 1.5*iqr
    uf= q3+ 1.5*iqr
    outliers= df[(df[col]<lf) | (df[col]>uf)]
    print("Percentage of outliers in", col ,outliers.shape[0]*100/df.shape[0])
    fig, ax= plt.subplots(1,2, figsize=(6,4))
    ax[0].boxplot(df[col])
    ax[1].hist(df[col])
    plt.suptitle(col)
    plt.show()
    df.loc[(df[col]<lf) | (df[col]>uf), col]= df[col].median()
for i in num:
  outlier_detection_treatment(i)
```

```python
spec_char = ['_______', '', '!@9#%8', 'NM', '_']

for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = df[column].replace(spec_char, np.nan)
df.dropna(inplace=True)
df
```

```python
df = df.reset_index()
```

```python
from sklearn.preprocessing import OneHotEncoder
oheMnth= OneHotEncoder(sparse_output= False)
np_Mnth= oheMnth.fit_transform(df[["Month"]])
df_Mnth= pd.DataFrame(np_Mnth,columns= oheMnth.categories_ )
df_Mnth.drop('January', axis= 1, inplace=True)
```

```python
oheOccu= OneHotEncoder(sparse_output= False)
np_Occu= oheOccu.fit_transform(df[["Occupation"]])
df_Occu= pd.DataFrame(np_Occu,columns= oheOccu.categories_ )
df_Occu.drop('Scientist', axis= 1, inplace=True)
```

```python
ohePay= OneHotEncoder(sparse_output= False)
np_Pay= ohePay.fit_transform(df[["Payment_Behaviour"]])
df_Pay= pd.DataFrame(np_Pay,columns= ohePay.categories_ )
df_Pay.drop('Low_spent_Small_value_payments', axis= 1, inplace=True)
```

```python
df['Credit_Mix'].replace(to_replace = ['Good', 'Standard', 'Bad'], value = [2, 1, 0], inplace = True)
df['Credit_Score'].replace(to_replace = ['Good', 'Standard', 'Poor'], value = [2, 1, 0], inplace = True)
df['Payment_of_Min_Amount'].replace(to_replace = ['No', 'Yes'], value= [0, 1], inplace = True)
```

```python
X= df.drop(['Month', 'Occupation', 'Payment_Behaviour', 'Credit_Score','index'], axis= 1)
y= df['Credit_Score']
X= pd.concat([X, df_Mnth, df_Occu, df_Pay], axis=1)
X
```

```python
!pip install tensorflow
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
```

```python
# Convert X to NumPy array if it's a DataFrame with mixed column types
X_np = X.to_numpy()  # Avoids errors from mixed column names
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_np)
```

```python

# Convert labels to categorical (for 3-class classification)
y_encoded = to_categorical(y, num_classes=3)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.3, random_state=0)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

# Define the model
model = Sequential()
model.add(Dense(64, input_shape=(X_train.shape[1],), activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(y_encoded.shape[1], activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model with 100 epochs
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
optimizer = Adam(learning_rate=0.0001)
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_accuracy', save_best_only=True, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(X_train, y_train, epochs=200, batch_size=2000, verbose=1, validation_split=0.2,
                    callbacks=[checkpoint, early_stopping])
```

```python
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'][:25], label='train loss')
plt.plot(history.history['val_loss'][:25], label='val loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.load_weights('best_model.keras')
y_pred_train = model.predict(X_train)
y_pred_train = np.argmax(y_pred_train, axis=1)
y_train = np.argmax(y_train, axis=1)

y_pred_test = model.predict(X_test)
y_pred_test = np.argmax(y_pred_test, axis=1)
y_test = np.argmax(y_test, axis=1)

# Accuracy
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Train Accuracy: {train_accuracy}")
print(f"Test Accuracy: {test_accuracy}")
```
